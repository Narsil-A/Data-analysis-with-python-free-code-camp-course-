```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np

def draw_plot():
    # Read data from file
    
    df = pd.read_csv('epa-sea-level.csv')

    df.head()


    # Create scatter plot

    # Creating a scatter plot using the 'Year' column as the x-axis and the 'CSIRO Adjusted Sea Level' column as the y-axis
    plt.figure(figsize=(10, 6))
    plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], c='blue', label='CSIRO Adjusted Sea Level')
    plt.xlabel('Year')
    plt.ylabel('CSIRO Adjusted Sea Level')
    plt.title('Scatter Plot of Global Average Sea Level Change')
    plt.legend()
    plt.show()

    # Calculate the line of best fit
    slope, intercept, r_value, p_value, std_err = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
    # Create the line of best fit
    years_extended = np.arange(1880, 2051, 1)
    line_best_fit = slope * years_extended + intercept

    # Create the scatter plot
    plt.figure(figsize=(12, 6))
    plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], label='Data')
    
    # Plot the line of best fit
    plt.plot(years_extended, line_best_fit, 'r', label='Line of Best Fit')

    plt.xlabel('Year')
    plt.ylabel('CSIRO Adjusted Sea Level (inches)')
    plt.title('Global Average Sea Level Change')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Creating scatter plot
    plt.figure(figsize=(12, 6))
    plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], label="Data Points")

    # Line of best fit using all data
    slope, intercept, r_value, p_value, std_err = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
    years_extended = list(range(1880, 2051))
    plt.plot(years_extended, [intercept + slope*x for x in years_extended], 'b', label="Best Fit Line (All Years)")

    # Line of best fit using data from year 2000 to most recent
    df_2000_onwards = df[df['Year'] >= 2000]
    slope_2000, intercept_2000, *_ = linregress(df_2000_onwards['Year'], df_2000_onwards['CSIRO Adjusted Sea Level'])
    years_2000_onwards = list(range(2000, 2051))
    plt.plot(years_2000_onwards, [intercept_2000 + slope_2000*x for x in years_2000_onwards], 'r', label="Best Fit Line (2000 Onwards)")

    # Labels and Title
    plt.xlabel("Year")
    plt.ylabel("Sea Level (inches)")
    plt.title("Rise in Sea Level")
    plt.legend()

    # Show the plot
    plt.show()
    plt.savefig('sea_level_plot.png')
    return 
draw_plot()
```
